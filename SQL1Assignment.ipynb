### Read the following data set:

```python
import pandas as pd
import numpy as np
```

Create a sql db from adult dataset and name it sqladb

```python
sqladb = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',header=None,
                 names=['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','yearly_income']
                     ,delimiter=',',skipinitialspace=True)
```

```python
#Display the sample rows
sqladb.head(4)
```

```python
from pandasql import sqldf
```

```python
#define the global function, as managing local and global environment might be confusing at times
pysqldf = lambda q: sqldf(q, globals())
```

 <B> 1. Select 10 records from the adult sqladb</B>

```python
pysqldf("SELECT * FROM sqladb LIMIT 10;")
```

<B> 2. Show me the average hours per week of all men who are working in private sector </B>

```python
pysqldf("SELECT AVG([hours-per-week]) as 'Avg_Hrs_week' FROM sqladb WHERE  workclass = 'Private' AND sex='Male';")
```

<B> 3. Show me the frequency table for education, occupation and relationship, separately </B>

```python
pysqldf("SELECT education,COUNT(*)  FROM sqladb GROUP BY education ;")
```

```python
pysqldf("SELECT occupation,COUNT(*)  FROM sqladb GROUP BY occupation ;")
```

```python
pysqldf("SELECT relationship,COUNT(*)  FROM sqladb GROUP BY relationship ;")
```

<B> 5. What is the average, minimum and maximum age group for people working in
different sectors</B>

```python

pysqldf("""SELECT workclass as "sector",
                AVG(age) as "Average_Age",
                MIN(age) as "Min_Age",
                MAX(age) as "Max_Age" 
            FROM sqladb 
            GROUP BY workclass """)
```

<B> 6. Calculate age distribution by country </B>

```python
pysqldf("""SELECT [native-country] as "country",
                AVG(age) as "Average_Age",
                MIN(age) as "Min_Age",
                MAX(age) as "Max_Age", 
                COUNT(*) as 'Age_Distribution'
            FROM sqladb 
            GROUP BY [native-country] """)
```

<B>7. Compute a new column as 'Net-Capital-Gain' from the two columns 'capital-gain' and
'capital-loss'</B>

```python
pysqldf("""SELECT *, [capital-gain] -[capital-loss] as "Net-Capital-Gain"             
            FROM sqladb 
            """)
```
